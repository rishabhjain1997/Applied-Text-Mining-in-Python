## Part 2 - Company Emails

For the second part of this assignment you will be workking with a company's email network where each node corresponds to a person at the company, and each edge indicates that at least one email has been sent between two people.

The network also contains the node attributes `Department` and `ManagementSalary`.

`Department` indicates the department in the company which the person belongs to, and `ManagementSalary` indicates whether that person is receiving a management position salary.

```python
G = nx.read_gpickle('email_prediction.txt')

print(nx.info(G))
```

```python
from sklearn.preprocessing import OneHotEncoder
df = pd.DataFrame(G.nodes())
df['Department'] = pd.Series(nx.get_node_attributes(G, 'Department'))
df['ManagementSalary'] = pd.Series(nx.get_node_attributes(G, 'ManagementSalary'))
df = df.set_index(0)
df['clustering'] = pd.Series(nx.clustering(G))
df['degree_centrality'] = pd.Series(nx.degree_centrality(G))
df['closeness_centrality'] = pd.Series(nx.closeness_centrality(G, normalized = True))
df['betweenness_centrality'] = pd.Series(nx.betweenness_centrality(G, normalized = True, endpoints = False))
df['pagerank'] = pd.Series(nx.pagerank(G, alpha = 0.8))
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
enc_df = pd.DataFrame(enc.fit_transform(df[['Department']]).toarray())
df = df.join(enc_df)
df.head()
```

```python
df = df.drop('Department', axis = 1)
df.head()
```

```python
test = df[df['ManagementSalary'].isnull()]
train = df[df['ManagementSalary'].notnull()]
y_train = train['ManagementSalary']
X_train = train.drop('ManagementSalary', axis = 1)
X_test = test.drop('ManagementSalary', axis = 1)
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier().fit(X_train, y_train)
clf.n_layers_
```

```python
clf.predict_proba_(X_test)
```
