```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
G = nx.karate_club_graph()
G = nx.convert_node_labels_to_integers(G, first_label = 1)
```

```python
degrees = G.degree()
degree_values = sorted(set(degrees.values()))
number_of_nodes = len(list(degrees.values()))
```

```python
histogram = [list(degrees.values()).count(degree_value)/number_of_nodes for degree_value in degree_values]
plt.bar(degree_values, histogram)
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.show()
```

```python
G = nx.generators.fast_gnp_random_graph(30, 0.5, directed = True)
in_degrees = G.in_degree()
in_degree_values = sorted(set(in_degrees.values()))
number_of_nodes = nx.number_of_nodes(G)
histogram = [list(in_degrees.values()).count(in_degree_value)/number_of_nodes for in_degree_value in in_degree_values]
plt.bar(in_degree_values, histogram)
plt.xlabel('In Degree')
plt.ylabel('Fraction of Nodes')
plt.show()
```

```python
G = nx.barabasi_albert_graph(1000000, 1)
degrees = G.degree()
degree_values = sorted(set(degrees.values()))
histogram = [list(degrees.values()).count(degree_value) for degree_value in degree_values]
plt.plot(degree_values, histogram, 'o')
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.yscale('log')
plt.xscale('log')
plt.show()
```
